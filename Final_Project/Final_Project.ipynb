Признак _Healthcare_1_ - удаляем, слишком много пропусков.

```python
data.drop('Healthcare_1', axis=1, inplace=True)
```

Видим несколько выбросов в столбцах _LifeSquare_, _Square_, _Rooms_ и _HouseYear_.    
Исправляем.

```python
data.loc[data['HouseYear'] > 2020]
```

```python
# 2005 - 2011? Возьмем среднее.
data.loc[data['HouseYear'] == 20052011, 'HouseYear'] = 2008
```

```python
# 4968? Скорее, 1968.
data.loc[data['HouseYear'] == 4968, 'HouseYear'] = 1968
```

```python
data.loc[data['Rooms'] > 6]
```

```python
# Судя по площади, скорее это 2-х комнатная квартира
data.loc[data['Rooms'] == 10, 'Rooms'] = 2
```

```python
# Похоже на однушку
data.loc[data['Rooms'] == 19, 'Rooms'] = 1
```

```python
# Удалим комнаты с количеством комнат 0, т.к. не хочу их править руками
rem_index = data.loc[data['Rooms'] < 1].index
data.drop(rem_index, inplace=True)
```

```python
mean_square_per_room = pd.DataFrame(data['Square'] / data['Rooms'], columns=['mean'])
mean_square_per_room = mean_square_per_room['mean'].mean()
mean_square_per_room
```

```python
# Исправим слишком большую и маленькую площадь
def prepare_square(df):
    df.loc[df['Square'] > (df['Rooms'] * (mean_square_per_room + 10)), 'Square'] = df['Rooms'] * (mean_square_per_room + 10)
    df.loc[df['Square'] < (df['Rooms'] * (mean_square_per_room - 10)), 'Square'] = df['Rooms'] * (mean_square_per_room - 10)
    return df
```

```python
# Если жилая площадь NaN или больше общей площади, надо это исправлять
def prepare_lifesquare(df, df_source):
    df['LifeSquare'] = df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    df.loc[df['LifeSquare'] > df['Square'], 'LifeSquare'] = df['Square'] - df['KitchenSquare'].mean()
    return df
```

```python
data = prepare_square(data)
data = prepare_lifesquare(data, data)
```

```python
target = np.array(data['Price'])
```

```python
features = ['DistrictId', 'Rooms', 'Square', 'LifeSquare', 'Floor', 'HouseYear', 'Ecology_1', 'Social_1', 'Social_2', 'Social_3', 'Helthcare_2', 'Shops_1']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(data[features], target, random_state=42)
```

```python
parameters = [{'n_estimators': [300, 350, 400],
               'max_features': np.arange(3, 7),
               'max_depth': np.arange(7, 10)}]
```

```python
reg = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                   param_grid=parameters,
                   scoring='r2',
                   cv=5)
```

```python
%%time
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train), reg.score(X_valid, y_valid)
```

```python
reg.best_params_
```
