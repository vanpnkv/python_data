```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
data.info()
```

```python
data.describe()
```

```python
data.head()
```

```python
data.drop('Healthcare_1', axis=1, inplace=True)
```

```python
data.info()
```

```python
data.loc[data['HouseYear'] > 2020]
```

```python
data.loc[data['HouseYear'] == 20052011, 'HouseYear'] = 2008
data.loc[data['HouseYear'] == 4968, 'HouseYear'] = 1968
```

```python
data.loc[data['Rooms'] > 6]
```

```python
data.loc[data['Rooms'] == 10, 'Rooms'] = 2
data.loc[data['Rooms'] == 19, 'Rooms'] = 1
```

```python
data.loc[data['Square'] > 150]
```

```python
def prepare_square(df):
    df.loc[df['Square'] > (df['Rooms']*40 + 20), 'Square'] = df['Rooms']*40 + 20
    df.loc[df['Square'] < (df['Rooms']*15), 'Square'] = df['Rooms']*15
    return df
```

```python
data.loc[data['Rooms'] == 0, 'Rooms'] = 1
data = prepare_square(data)
```

```python
data.describe()
```

```python
def fillna_lifesquare(df, df_source):
    df['LifeSquare'] = df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    return df
```

```python
data = fillna_lifesquare(data, data)
```

```python
data.info()
```

```python
def join_stats1(df, stats1, source_df, mode = 'train'):
    df = pd.merge(df, stats1, on = ['DistrictId', 'Rooms'], how='left')
    if mode == 'test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
stats1 = data.groupby(['DistrictId', 'Rooms'], as_index=False)[['Price']].mean().rename(columns={'Price': 'mean_price'})
```

```python
data = join_stats1(data, stats1, data)
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = data[['Rooms', 'Square', 'LifeSquare', 'mean_price']]
target = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression as LR, Ridge
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
model1 = LR()
```

```python
%%time
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print(mean_squared_error(y_test, y_pred), r2_score(y_test, y_pred))
```

```python
model2 = Ridge(alpha=0.05)
```

```python
%%time
model2.fit(X_train, y_train)
```

```python
y_pred_2 = model2.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred_2), r2_score(y_test, y_pred_2))
```

```python
model3 = RF(n_estimators=100, max_depth=10, min_samples_leaf=5)
```

```python
%%time 
model3.fit(X_train, y_train)
```

```python
y_pred_3 = model3.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred_3), r2_score(y_test, y_pred_3))
```
