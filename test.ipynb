Бенчмарк должен показать характеристику производительности вашего алгоритма для массивов случайных чисел в зависимости от длины (необходимо продемонстрировать график).   
Длина может варьироваться в диапазоне от 1 до 2^24 элементов.   
На дальнейшем интервью мы спросим, почему график получился таким каким он у вас получится, и если он будет отличаться от графика теоретической  O-сложности, то почему.

```python
import numpy as np
```

```python
# Bubble sort
def sort(array, coArray):
    for k in range(len(array)):
        for i in range(len(array) - k - 1):
            if array[i] > array[i+1]:
                array[i], array[i+1] = array[i+1], array[i]
                coArray[i], coArray[i+1] = coArray[i+1], coArray[i]
```
